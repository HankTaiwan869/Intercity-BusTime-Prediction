```python
import polars as pl
import plotly.express as px
from datetime import datetime

pl.Config.set_tbl_rows(30)
```

```python
df_complete = pl.read_csv("raw_data/sample.csv")
df_complete.glimpse()
df_complete.columns
```

```python
df_complete.head()
```

```python
# Prepare df of a specific route

TARGET_ROUTEID = 1728

df = (
    df_complete.filter(
        pl.col("RouteID") == TARGET_ROUTEID,
        pl.col("TripStartTimeType") == "實際發車時間",
    )
    .with_columns(
        pl.col("GPSTime")
        .str.replace(r"\+08:00", "")
        .str.to_datetime(format="%Y-%m-%dT%H:%M:%S")
        .alias("Time"),
    )
    .drop(
        [
            "GPSTime",
            "OperatorID",
            "OperatorNo",
            "RouteUID",
            "RouteID",
            "OperatorID",
            "OperatorNo",
            "RouteUID",
            "RouteID",
            "StopUID",
            "StopID",
            "TripStartTime",
            "TransTime",
            "SrcRecTime",
            "SrcTransTime",
            "SrcUpdateTime",
            "UpdateTime",
            "RouteNameZh_tw",
            "RouteNameEn",
            "SubRouteUID",
            "SubRouteID",
            "SubRouteNameZh_tw",
            "SubRouteNameEn",
            "StopNameEn",
            "StopSequence",
            "MessageType",
            "TripStartTimeType",
        ]
    )
)
df.glimpse()
```

```python
fig = px.bar(
    df.filter(pl.col("Direction") == "去程")
    .group_by(["StopNameZh_tw", "A2EventType"])
    .len()
    .sort(["StopNameZh_tw", "A2EventType"]),
    y="StopNameZh_tw",
    x="len",
    color="A2EventType",
)
fig.show(renderer="browser")
```
