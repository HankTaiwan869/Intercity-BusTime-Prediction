```python
import lightgbm as lgb
import pandas as pd
from constants import MODELS
from datetime import time
```

```python
# Load native format
model = lgb.Booster(model_file=MODELS / "model_1728.txt")
```

```python
def predict(model: lgb.Booster, target_time: time, day_of_week: str) -> float:
    minutes_from_midnight = target_time.hour * 60 + target_time.minute
    day_of_week_categories = ["Thu", "Fri", "Sat", "Sun", "Mon", "Tue", "Wed"]

    X = pd.DataFrame(
        [
            {
                "Minutes_from_midnight": minutes_from_midnight,
                "Day_of_week": day_of_week,
            }
        ]
    )

    X["Day_of_week"] = pd.Categorical(
        X["Day_of_week"], categories=day_of_week_categories
    )
    minutes_from_midnight = target_time.hour * 60 + target_time.minute
    return model.predict(X).item()
```

```python
for hour in range(1, 24):
    print(f"{hour} - {predict(model, time(hour, 0, 0), 'Fri'):.2f} min")
```
