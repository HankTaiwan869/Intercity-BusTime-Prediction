# Singe Route Predictions
## Context
- The lightGBM model is trained and is ready for deployment (DEMO version)
## Goal
- Figure out how to make predictions for deployment in Streamlit DEMO

```python
import lightgbm as lgb
import pandas as pd
from constants import DEMO_FOLDER, DAY_CATEGORIES
from datetime import time
```

```python
# Load native format
model = lgb.Booster(model_file=DEMO_FOLDER / "demo_model_1728.txt")
```

```python
def predict(model: lgb.Booster, target_time: time, day_of_week: str) -> float:
    minutes_from_midnight = target_time.hour * 60 + target_time.minute

    X = pd.DataFrame(
        [
            {
                "Minutes_from_midnight": minutes_from_midnight,
                "Day_of_week": day_of_week,
            }
        ]
    )

    X["Day_of_week"] = pd.Categorical(X["Day_of_week"], categories=DAY_CATEGORIES)
    minutes_from_midnight = target_time.hour * 60 + target_time.minute
    return model.predict(X).item()
```

```python
for hour in range(1, 24):
    print(f"{hour} - {predict(model, time(hour, 0, 0), 'Fri'):.2f} min")
```
