# 04 Modeling Training & Hyperparameters Tuning

```python
import lightgbm as lgb
from constants import PROCESSED_DATA_FOLDER
```

```python
train = lgb.Dataset(PROCESSED_DATA_FOLDER / "lightgbm_train.bin")
test = lgb.Dataset(PROCESSED_DATA_FOLDER / "lightgbm_test.bin")
```

```python
params = {
    "objective": "regression",
    "metric": "rmse",
    "verbosity": -1,
    "num_leaves": 31,
    "max_depth": -1,
    "min_data_in_leaf": 20,
    "learning_rate": 0.05,
    "feature_fraction": 0.9,  # colsample
    "bagging_fraction": 0.8,  # subsample
    "bagging_freq": 5,
    "lambda_l1": 0.0,
    "lambda_l2": 0.0,
}

model = lgb.train(
    params=params,
    train_set=train,
    num_boost_round=100,
    valid_sets=[train, test],
    valid_names=["train", "test"],
    callbacks=[
        lgb.early_stopping(stopping_rounds=50),
        lgb.log_evaluation(period=100),
    ],
)
```
